```python
from transformers import GPT2LMHeadModel, GPT2Tokenizer

# Load the GPT-2 tokenizer and model
tokenizer = GPT2Tokenizer.from_pretrained("gpt2-large")
model = GPT2LMHeadModel.from_pretrained("gpt2-large")

print("GPT-2 large model and tokenizer loaded successfully!")
```

```python
# Load the GPT-2 tokenizer and model
tokenizer_gpt2 = GPT2Tokenizer.from_pretrained("gpt2")
model_gpt2 = GPT2LMHeadModel.from_pretrained("gpt2")

print("GPT-2 model and tokenizer loaded successfully!")
```

```python
# Define a prompt
prompt = "Once upon a time,"

# Encode the prompt and generate text
input_ids = tokenizer.encode(prompt, return_tensors="pt")
output = model.generate(input_ids, max_length=100, num_return_sequences=1, no_repeat_ngram_size=2)

# Decode and print the generated text
generated_text = tokenizer.decode(output[0], skip_special_tokens=True)
print(generated_text)
```

```python
system_prompt = '''
### **Role**

You are a python expert which can write any python program and can answer any question related to python language.

### **Task**

Your task to analyze user's question in natural language, verify if the question is related to python, if yes, give proper answer with supported code if required, if not, just say "Sorry, I am a Python only expert, can't help with any other concerns."

### **Constraints**

1. You should strictly not answer any question which is **not** related to python.
2. If user asks any programming question without mentioning the language, by default consider it to be python and anser it accordingly.

### **Examples**

**Example 1: Theoritical Question**
**User Input:**  Explain Global Interpreter Lock (GIL) in Python.
**Expected Output: **
  * GIL ensures only one thread runs Python bytecode at a time (even on multi-core CPUs).

  * Good for I/O-bound tasks, but limits CPU-bound multithreading.

  * Workaround: use multiprocessing instead of threads.

**Example 2: Coding Question**
**User Input:** How do you swap two variables without using a temporary variable?
**Expected Output: **
x, y = 10, 20
x, y = y, x
print(x, y)  # 20 10

**Example 3: Question not related to Python**
**User Input:** What is the capital of France?
**Expected Output: ** Sorry, my expertise is limited to Python, and I’m unable to help with other areas..

Now assist users for their quesries.

**User Query: **
'''
```

```python
user_query = 'What is the capital of Japan?'
```

```python
prompt = """You are an assistant that ONLY answers questions about Python programming.
- If the user asks something about Python (syntax, libraries, debugging, concepts), provide a clear and helpful answer.
- If the user asks about anything else, respond with:
  "Sorry, I can only help with Python-related questions."
Do not break this rule under any circumstance.

User: {}
Assistant:""".format(user_query)
```

```python
user_query = 'What is the capital of Japan?'

prompt = system_prompt + user_query

# Encode the prompt and generate text
input_ids = tokenizer.encode(prompt, return_tensors="pt")

# Add attention mask and set pad_token_id
attention_mask = input_ids.new_ones(input_ids.shape)
pad_token_id = tokenizer.eos_token_id

output = model_gpt2.generate(input_ids,
                        max_length=1000,
                        num_return_sequences=1,
                        no_repeat_ngram_size=2,
                        attention_mask=attention_mask,
                        pad_token_id=pad_token_id)

# Decode and print the generated text
generated_text = tokenizer.decode(output[0], skip_special_tokens=True)

# Extract only the generated response by finding the end of the prompt
prompt_end_index = generated_text.find(user_query) + len(user_query)
extracted_response = generated_text[prompt_end_index:].strip()

print(extracted_response)
```
